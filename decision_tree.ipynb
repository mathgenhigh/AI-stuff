```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from statistics import mean, stdev
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
data = load_iris()
```

```python
X = data.data
Y = data.target
```

```python
clf = DecisionTreeClassifier(random_state=1)
```

Task 8: a decision tree is generally considered "good enough" when its validation accuracy is significantly higher than a random guess and comparable to more complex models, without showing signs of overfitting.

Overall the best results are happening on the 0.25 & 0.75 split ratios

```python
accuracy = []
test_sizes = [0.1, 0.25, 0.5, 0.75, 0.9]

for i in range(1, 10 + 1):
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=i)

    clf.fit(X_train, Y_train)

    Y_pred = clf.predict(X_test)
    Y_proba = clf.predict_proba(X_train)
    accuracy.append(accuracy_score(Y_test, Y_pred))

for idx, val in enumerate(accuracy):
    print(f"Accuracy at {idx + 1}: {val}")
```

```python
mean_val = mean(accuracy)
stdev(accuracy)
print(f"Mean: {mean_val}")
print(f"Standard deviation: {stdev(accuracy)}")
```

```python
accuracy_2 = []
test_sizes = [0.1, 0.25, 0.5, 0.75, 0.9]

for size in test_sizes:
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=size, random_state=42)

    clf.fit(X_train, Y_train)

    Y_pred = clf.predict(X_test)
    Y_proba = clf.predict_proba(X_train)
    accuracy_2.append(accuracy_score(Y_test, Y_pred))

for idx, val in enumerate(accuracy_2):
    print(f"Accuracy at {idx + 1}: {val}")
```

```python
plt.plot(test_sizes, accuracy_2)
```

```python
clf2 = tree.DecisionTreeClassifier(criterion='entropy')
clf2 = clf2.fit(X, Y)
tree.plot_tree(clf2)
plt.show()
```
